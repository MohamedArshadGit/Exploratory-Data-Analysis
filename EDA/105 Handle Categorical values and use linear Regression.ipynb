```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Startups_Expense.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['State'].unique()
```

```python
state=pd.get_dummies(df['State'],drop_first=True).astype(int)
```

```python
df=pd.concat([state,df],axis=1)
```

```python
df.head()
```

```python
df.drop(columns=['State'],inplace=True)
```

```python
df
```

```python
X=df.iloc[:,:-1]
```

```python
y=df.iloc[:,5]
```

```python
from sklearn.model_selection import train_test_split

```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
y_pred=linear.predict(X_test)
```

```python
from sklearn.metrics import explained_variance_score,r2_score,mean_absolute_error,mean_squared_error
```

```python
print("evs",explained_variance_score(y_test,y_pred))
print("r2_score",r2_score(y_test,y_pred))
print("mae",mean_absolute_error(y_test,y_pred))
print("mse",mean_squared_error(y_test,y_pred))
print('rmse:',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
data = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

sns.regplot(data=data, x='Actual', y='Predicted')
plt.show()
```

```python
sns.regplot(x=y_test,y=y_pred,line_kws={'color':'red'},scatter_kws={'color':'blue'})
```

```python
plt.scatter(y_test, y_pred, color='b', label='Data') 
plt.plot([np.min(y_test),np.max(y_test)],[np.min(y_pred),np.max(y_pred)], color='r', label='Line of Regression')
#plt.scatter(0, intercept, color='y', label='Y_Intercept')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
